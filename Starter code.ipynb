```python
import pandas as pd
```

```python
citibikes_data = ["resources/202208.csv", "resources/202209.csv", "resources/202210.csv", "resources/202211.csv", "resources/202212.csv"]
```

```python
# Create an empty list to store each dataframe
citibikes_list = []

# Loop through each file in the list
for data in citibikes_data:
    # Read the file into a dataframe
    citibikes = pd.read_csv(data)

    # Append the dataframe to the list
    citibikes_list.append(citibikes)

# Concatenate all dataframes in the list into one dataframe
citibikes_df = pd.concat(citibikes_list)

# Show dataframe
citibikes_df
```

```python
citibikes_df.columns
```

```python
citibikes_df.isnull().values.any()
```

```python
citibikes_df.dropna(inplace = True)
citibikes_df
```

```python
citibikes_df.to_csv("resources/citibikes_combine_2.csv", index=False)
```
